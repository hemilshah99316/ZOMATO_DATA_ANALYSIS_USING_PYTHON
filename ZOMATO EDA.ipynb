```python
"""What does the heatmap show?

The heatmap visually represents where missing data (NaN values) are located in the zomato DataFrame.
Each cell in the heatmap corresponds to a cell in the DataFrame: cells with missing data will be colored differently (according to the 'viridis' colormap) than those without missing data.
Since the yticklabels is set to False, you won't see the row labels, making the heatmap less cluttered if the DataFrame has many rows.
The cbar=False means there's no color bar showing what the colors represent, but typically, the heatmap's colors will indicate whether a value is missing (True) or not (False).
"""
plt.figure(figsize=(15, 10))  # Increase the figure size
sns.heatmap(zomato.isnull(),yticklabels= False,cbar= False,cmap='viridis')
plt.show()
```

```python
country=pd.read_excel(r"C:\Users\hemil\OneDrive\Desktop\Data Analyst\EDA PYTHON\Krish Naik ZOMATO EDA\Country-Code.xlsx")
country.head()
```

```python
## How many records belong to which country(segregation of records based on Country)

country_names=final_zomato.Country.value_counts().index ## this is for developing the pie chart
country_values=final_zomato.Country.value_counts().values ## this is for developing the pie chart
## top 3 country where zomato sells
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%') ##Making the pie chart

##Answer:Zomato maximum bussiness comes from India
```

```python
## Relationship between ratings,colors,text using group by function
final_zomato.groupby(['Aggregate rating','Rating color','Rating text']).size()

```

```python
"""Answer:
    rating = 0.0-white-not rated
    rating = 1.8 to 2.4-red-poor
    rating = 2.5-3.4-orange-average
    rating = 3.5-3.9-yellow-good
    rating = 4-4.4-green-very good
    rating = 4.4-4.9-Dark-green-excellent
Rating is between 0 to 5
Many guys have not rated the zomato application
You can also derive many more things from the data"""
```

```python
##This is necessary for making charts out of it 
Ratings=final_zomato.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
Ratings

```

```python
"""
3 reset_index():
Flattening the Result: After using groupby and size, the result is a Series with a multi-index (one index for each of the group-by columns). The .reset_index() function converts this Series back into a DataFrame by moving the indices (which are the grouping columns) back into regular columns.
4 rename(columns={0: 'Rating Count'}):
Renaming the Column: The .size() function creates a new column in the DataFrame that holds the count of each group. By default, this column is unnamed (indexed as 0). The .rename(columns={0: 'Rating Count'}) part renames this column to 'Rating Count' for clarity.
"""
```

```python
# Assuming Ratings is your DataFrame and it has already been created
plt.figure(figsize=(12, 6))
sns.barplot(x='Aggregate rating', y='Rating Count', data=Ratings)
# Show the plot
plt.show()
```

```python
# More sophisticated chart(details are same as above)
# Plot the barplot with the filtered data
plt.figure(figsize=(12, 6))
sns.barplot(x='Aggregate rating', y='Rating Count',hue='Rating color', data=Ratings,palette=['black','red','orange','yellow','green','darkgreen'])
plt.show()
```

```python
"""
Answer:
1:Not Rated counts are very high
2:Maximum number of rating are between 2.5 to 3.4
"""
```

```python
### Create a pie chart for cities distribution(Cities where most orders are placed )
top_cities=final_zomato[['City']].groupby(['City']).size().sort_values(ascending=False).head().reset_index().rename(columns={0:'Orders_Placed'})
top_cities
```

```python
plt.figure(figsize=(10, 7))
plt.pie(top_cities['Orders_Placed'], labels=top_cities['City'], autopct='%1.1f%%')

# Add title
plt.title('Top Cities by Order Placement')
plt.show()
```

```python
## find the Top 10 Cuisines
Top_Cuisines=final_zomato[['Cuisines']].groupby(['Cuisines']).size().sort_values(ascending=False).reset_index().rename(columns={0:'Most loved Cuisines'}).head(10)
Top_Cuisines
```

```python
"""
Answer:Most loved cuisines are North Indian,Chinese,Mughlai
"""
```

```python
plt.pie(Top_Cuisines['Most loved Cuisines'],labels=Top_Cuisines['Cuisines'],autopct='%1.1f%%')
# Add title
plt.title('Top Cuisines by Order Placement')
plt.show()
```
